# General Knowledge:
* In Each NBA Season we have 82 games
* In basketball, a double-double or triple-double is when a player accumulates ten or more (a double-digit total) in two or three of five statistical categories—points, rebounds, assists, steals, and blocked shots—in a single game. The most common double-double combination is points and rebounds, followed by points and assists. The most common way to achieve a triple-double is through points, rebounds, and assists.

```python
# Imports
import numpy as np
import pandas as pd

%matplotlib inline

import matplotlib
import matplotlib.pyplot as plt
```

```python
totals_stats_df = pd.read_csv('../dataset/totals_stats.csv')
sal_df = pd.read_csv('../dataset/salaries.csv')
```

```python
totals_stats_df.head(5)
```

```python
# Tamperd data - to float
# Tamperd data - normelized (values not grater the 1000) 
sal_df['Salary'] = sal_df['Salary'].str.replace("$","").astype(float).div(other = 1000000, fill_value = 0).round(2) 
sal_df.head(5)
```

```python
labron_points_stats =  totals_stats_df[totals_stats_df.Player.eq("Lebron James")].drop_duplicates(['Season'])
kobe_points_stats =  totals_stats_df[totals_stats_df.Player.eq("Kobe Bryant")].drop_duplicates(['Season'])
michael_points_stats =  totals_stats_df[totals_stats_df.Player.eq("Michael Jordan")].drop_duplicates(['Season'])
lb_main_stats = labron_points_stats[['Season','PTS','AST','BLK','ORB','DRB','STL']]
kb_main_stats = kobe_points_stats[['Season','PTS','AST','BLK','ORB','DRB','STL']]
mj_main_stats = michael_points_stats[['Season','PTS','AST','BLK','ORB','DRB','STL']]

lb_sal = sal_df[sal_df.Player.eq("Lebron James")]
kb_sal = sal_df[sal_df.Player.eq("Kobe Bryant")]
mj_sal = sal_df[sal_df.Player.eq("Michael Jordan")]
```

```python
def player_stats(df):
    # set width of bar
    barWidth = 0.15
 
    # set height of bar
    bars1 = df['PTS']
    bars2 = df['AST']
    bars3 = df['BLK']
    bars4 = df['ORB']
    bars5 = df['DRB']
    bars6 = df['STL']

    # Set position of bar on X axis
    r1 = np.arange(len(bars1))
    r2 = [x + barWidth for x in r1]
    r3 = [x + barWidth for x in r2]
    r4 = [x + barWidth for x in r3]
    r5 = [x + barWidth for x in r4]
    r6 = [x + barWidth for x in r5]

    # Make the plot
    plt.bar(r1, bars1, color='#7f6d5f', width=barWidth, edgecolor='white', label='Points Per Season')
    plt.bar(r2, bars2, color='#557f2d', width=barWidth, edgecolor='white', label='Asists Per Season')
    plt.bar(r3, bars3, color='#2d7f5e', width=barWidth, edgecolor='white', label='Blocks Per Season')
    plt.bar(r4, bars4, color='#084b8a', width=barWidth, edgecolor='white', label='O-ReBounds Per Season')
    plt.bar(r5, bars5, color='#97cbff', width=barWidth, edgecolor='white', label='D-ReBounds Per Season')
    plt.bar(r6, bars6, color='#fbcc7a', width=barWidth, edgecolor='white', label='Steals Per Season')

    # Add xticks on the middle of the group bars
    plt.xlabel('Season', fontweight='bold')
    plt.xticks( [r + barWidth for r in range(len(bars1))], df['Season'],rotation=90)

    # Add yLables
    plt.ylabel('ylabel', fontsize=16)

    # Set figure 
    fig_size = plt.rcParams["figure.figsize"]
    fig_size[0] = 12
    fig_size[1] = 6
    plt.rcParams["figure.figsize"] = fig_size


    # Create legend & Show graphic
    plt.legend()
    plt.show()


def player_sal(df):
    # set width of bar
    barWidth = 0.25
 
    # set height of bar
    bars1 = df['Salary']

    # Set position of bar on X axis
    r1 = np.arange(len(bars1))

    # Make the plot
    plt.bar(r1, bars1, color='#7f6d5f', width=barWidth, edgecolor='white', label='SAL Per Season')

    # Add xticks on the middle of the group bars
    plt.xlabel('Season', fontweight='bold')
    plt.xticks([r + barWidth for r in range(len(bars1))], df['Season'] ,rotation=90)

    # Set figure 
    fig_size = plt.rcParams["figure.figsize"]
    fig_size[0] = 12
    fig_size[1] = 6
    plt.rcParams["figure.figsize"] = fig_size

    # Create legend & Show graphic
    plt.legend()
    plt.show()
```
